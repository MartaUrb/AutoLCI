```python
#Create subdirectories for respective frames

def create_subdirectories(base_directory, prefix, numbers):
    # Ensure the base directory exists
    os.makedirs(base_directory, exist_ok=True)
    
    # List to keep track of created directories
    subdirs = []
    
    # Create subdirectories with prefix and numbers
    for number in numbers:
        dir_name = f"{prefix}{number}"
        path = os.path.join(base_directory, dir_name)
        
        # Check if the directory already exists before creating
        if not os.path.exists(path):
            os.makedirs(path)
            print(f"Directory created: {path}")
        else:
            print(f"Directory already exists: {path}")
        
        # Add to the list of all processed directories
        subdirs.append(path)
        
    return subdirs

# Example usage
prefix = filenameroot+'_frame'  # Prefix for the subdirectories
numbers = range(1, 4)  # Numbers 1, 2, 3

subdirs = create_subdirectories(motherdir+"/"+filenameroot, prefix, numbers)
print("sub-directories:", subdirs)

```

```python
def get_creation_times(motherdir, filenameroot, filelist):
    creation_times = {}
    for filename in filelist:
        # Construct the full path to the file
        file_path = os.path.join(motherdir, filenameroot, filename)
        
        # Get the file stats
        try:
            stat_info = os.stat(file_path)
            # Extract the creation time and convert to a readable format
            creation_time = datetime.datetime.fromtimestamp(stat_info.st_birthtime)
            creation_times[filename] = creation_time
        except AttributeError:
            # Handle the case where st_birthtime is not available
            creation_times[filename] = "Creation time not supported on this platform"
        except FileNotFoundError:
            # Handle the case where the file does not exist
            creation_times[filename] = "File not found"

    return creation_times

creation_times = get_creation_times(motherdir, filenameroot, filelist)
for file, time in creation_times.items():
    print(f"{file}: {time}")
```

```python
# Load the TIFF file
file_path = '/mnt/data/GV41_PL2_d02_7prc_01.tiff'
tiff_image = Image.open(file_path)

# Check how many frames or layers are there in the tiff file
frame_count = getattr(tiff_image, "n_frames", 1)

# Output the number of frames and the mode (to understand the structure of the file)
frame_count, tiff_image.mode





# Function to extract and save each frame in a separate folder
def extract_and_save_frames(tiff_img, num_frames, base_folder_path):
    for i in range(num_frames):
        # Set the directory for the current frame
        dir_path = os.path.join(base_folder_path, f"Image_{i+1}")
        os.makedirs(dir_path, exist_ok=True)
        
        # Select the frame
        tiff_img.seek(i)
        frame = tiff_img.copy()
        
        # Save the frame to the corresponding directory
        frame.save(os.path.join(dir_path, f"frame_{i+1}.png"))

# Base folder path for saving images
base_folder = '/mnt/data/extracted_images'

# Extract and save the frames
extract_and_save_frames(tiff_image, frame_count, base_folder)

# Provide paths to saved images for confirmation
output_directories = [f"/mnt/data/extracted_images/Image_{i+1}/frame_{i+1}.png" for i in range(frame_count)]
output_directories
```
