```python
import os as os
import tkinter as tk
from tkinter import filedialog
import tifffile as tiff
from PIL import Image
import numpy as np

#select directory with image files
scriptdir = os.path.dirname(os.getcwd()) # get the script directory
root = tk.Tk()
root.withdraw()
filedir = tk.filedialog.askdirectory(initialdir  = scriptdir) # manually select the directory with the files to rename
motherdir = os.path.dirname(filedir)
destdir = os.path.join(filedir,'concatanated')

filelist = [f for f in os.listdir(filedir) if f.endswith('.tiff') and '._' not in f] # make a list of all tsv files in data directory
filelist = [os.path.join(filedir, f) for f in filelist] #add path to file name
filelist.sort(key=os.path.getmtime) #sort files by time
```

```python
#create a list of times for timestamp
filenames = [os.path.basename(f) for f in filelist] #remove path from filename

from datetime import datetime

def extract_datetimes(file_names):
    # Define the datetime format from the file names
    datetime_format = '%Y%m%d_%H%M%S'

    # Initialize a list to store the extracted datetime objects
    extracted_datetimes = []

    # Process each file name
    for file_name in file_names:
        # Extract the datetime string based on the known length and position
        datetime_str = file_name[:15]
        # Convert the string to a datetime object
        date_time = datetime.strptime(datetime_str, datetime_format)
        # Append the datetime object to the list
        extracted_datetimes.append(date_time)

    return extracted_datetimes


# Extract datetimes from the file names
datetimes = extract_datetimes(filenames)
for dt in datetimes:
    print(dt)


def convert_to_absolute_time(datetimes, starting_point_str):
    # Parse the starting point datetime
    starting_point = datetime.strptime(starting_point_str, '%Y%m%d_%H%M%S')
    
    # Initialize a list to store the absolute time in HHMM format
    absolute_times = []
    
    # Process each datetime
    for dt in datetimes:
        # Calculate the difference in minutes
        total_minutes = int((dt - starting_point).total_seconds() // 60)
        # Convert minutes to days, hours and remaining minutes
        days = total_minutes // 1440
        hours = (total_minutes % 1440) // 60
        hours_nd = total_minutes // 60 #when no days included
        minutes = total_minutes % 60
        # Format the time in HHMM
        hhmm = f"{hours_nd:02d}:{minutes:02d}"
        dd_hh_mm = f"day{days:02d} {hours:02d}h{minutes:02d}"

        absolute_times.append(dd_hh_mm)
    
    return absolute_times


# Starting point
starting_point_str = '20240423_175500'  # Format: YYYYMMDD_HHMMSS

# Convert and print the absolute times
absolute_times = convert_to_absolute_time(datetimes, starting_point_str)
for time in absolute_times:
    print(time)
```

```python

def extract_and_combine_frames(tiff_paths, output_folder):
    # Make sure the output folder exists
    os.makedirs(output_folder, exist_ok=True)

    # Prepare a dictionary to hold lists of images for each frame index
    frames_dict = {0: [], 1: [], 2: []}

    # Process each tiff file
    for file_path in tiff_paths:
        with Image.open(file_path) as img:
            try:
                # Extract up to three frames
                for i in range(3):
                    img.seek(i)
                    # Append a copy of the frame to avoid reference issues
                    frames_dict[i].append(img.copy())
            except EOFError:
                # Fewer than three frames in this stack
                continue

    # Save new TIFF files containing the combined frames of each index
    for frame_index, images in frames_dict.items():
        # Create a multi-frame TIFF
        output_path = os.path.join(output_folder, f"combined_frame_{frame_index}.tiff")
        images[0].save(output_path, save_all=True, append_images=images[1:])

    print(f"Files saved in {output_folder}")


extract_and_combine_frames(filelist, destdir)
```

```python
## ADD TIMESTAMP TO STACKS
from PIL import Image, ImageDraw, ImageFont

def add_text_to_tiff_stack(input_file_path, output_file_path, texts):
    # Open the TIFF stack
    with Image.open(input_file_path) as img:
        # Create a list to hold the modified frames
        frames = []
        
        # Process each frame along with its corresponding text
        for i in range(len(texts)):
            try:
                img.seek(i)
                # Copy the frame to edit
                frame = img.copy()
                # Get a drawing context
                draw = ImageDraw.Draw(frame)
                
                # Define font
                try:
                    font = ImageFont.truetype("arial.ttf", size=40)  # Adjust size as needed
                except IOError:
                    font = ImageFont.load_default()
                
                # Define text position
                text_position = (frame.width - 200, frame.height - 50)  # Adjust position as needed
                
                # Draw text
                draw.text(text_position, texts[i], font=font, fill=(255, 255, 255))
                
                # Append the modified frame to the list
                frames.append(frame)
            except EOFError:
                break  # No more frames
        
        # Save the new TIFF stack with all modified frames
        frames[0].save(output_file_path, save_all=True, append_images=frames[1:], compression="tiff_deflate")

# Example usage
input_file_path = 'path/to/original/tiff_stack.tiff'
output_file_path = 'path/to/modified/tiff_stack.tiff'

add_text_to_tiff_stack(input_file_path, output_file_path, absolute_times)
```

```python
from PIL import Image, ImageDraw, ImageFont
import os

def extract_add_text_and_combine_frames(tiff_paths, output_folder, start_datetime_str):
    # Make sure the output folder exists
    os.makedirs(output_folder, exist_ok=True)

    # Prepare a dictionary to hold lists of images for each frame index
    frames_dict = {0: [], 1: [], 2: []}
    datetime_format = '%Y%m%d_%H%M%S'
    start_datetime = datetime.strptime(start_datetime_str, datetime_format)

    # Process each TIFF file
    for file_path in tiff_paths:
        with Image.open(file_path) as img:
            try:
                # Extract up to three frames
                for i in range(3):
                    img.seek(i)
                    frame = img.copy()

                    # Calculate absolute time text
                    current_datetime_str = os.path.basename(file_path)[:15]
                    current_datetime = datetime.strptime(current_datetime_str, datetime_format)
                    # Calculate the difference in minutes
                    total_minutes = int((current_datetime  - start_datetime).total_seconds() // 60)
                    # Convert minutes to days, hours and remaining minutes
                    days = total_minutes // 1440
                    hours = (total_minutes % 1440) // 60
                    hours_nd = total_minutes // 60 #when no days included
                    minutes = total_minutes % 60
                    # Format the time in DDHHMM or HHMM
                    #hhmm = f"{hours_nd:02d}:{minutes:02d}"
                    #dd_hh_mm = f"day{days:02d} {hours:02d}h{minutes:02d}"
                    absolute_time = f"day{days:02d} + {hours:02d}h{minutes:02d}"

                    # Get a drawing context and add text
                    draw = ImageDraw.Draw(frame)
                    try:
                        font = ImageFont.truetype("/Library/Fonts/Arial.ttf", size=100)
                    except IOError:
                        font = ImageFont.load_default()
                    text_position = (frame.width - 800, 100)  # Adjust position as needed
                    draw.text(text_position, absolute_time, font=font, fill=(255, 255, 255))

                    # Append the modified frame to avoid reference issues
                    frames_dict[i].append(frame)
            except EOFError:
                # Fewer than three frames in this stack
                continue

    # Save new TIFF files containing the combined frames of each index
    for frame_index, images in frames_dict.items():
        if images:  # Check if there are images to save
            output_path = os.path.join(output_folder, f"combined_frame_{frame_index}.tiff")
            images[0].save(output_path, save_all=True, append_images=images[1:], compression="tiff_deflate")
            print(f"Saved {output_path}")

# Example usage
starting_point = '20240423_175500'  # YYYYMMDD_HHMMSS format

extract_add_text_and_combine_frames(filelist, destdir, starting_point)
```
